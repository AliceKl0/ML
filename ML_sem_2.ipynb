```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class GradientDescent:
    def __init__(self, func, grad_func, learning_rate=0.01, max_iters=1000):
        self.func = func
        self.grad_func = grad_func
        self.learning_rate = learning_rate
        self.max_iters = max_iters
    
    def step(self, params):
        grad = self.grad_func(params)
        new_params = params - self.learning_rate * grad

        return new_params
    
    def minimize(self, init_params):
        params = init_params
        
        for i in range(self.max_iters):
            params = self.step(params)
            
            if i % 100 == 0:
                print(f"Итерация {i}: \nПараметры {params} \nЗначение ф-ции: {self.func(params)}\n")
        
        return params, self.func(params)
```

```python
def func_2d(params):
    x, y = params
    return x**2 + y**2

def grad_func_2d(params):
    x, y = params
    return np.array([2*x, 2*y])
```

```python
gd_2d = GradientDescent(func_2d, grad_func_2d, learning_rate=0.1, max_iters=1000)
```

```python
min_params_2d, min_value_2d = gd_2d.minimize(np.array([1.0, 1.0]))
print(f"Мин. знач. ф-ции: {min_value_2d} \nПри параметрах {min_params_2d}")
```

```python
def func_3d(params):
    x, y, z = params
    return x**2 + (y - 3) ** 2 + (z + 15) ** 2

def grad_func_3d(params):
    x, y, z = params
    return np.array([2*x, 2 * (y - 3), 2 * (z + 15)])
```

```python
gd_3d = GradientDescent(func_3d, grad_func_3d, learning_rate=0.1, max_iters=1000)
```

```python
min_params_3d, min_value_3d = gd_3d.minimize(np.array([1.0, 1.0, 1.0]))
print(f"Мин. знач. ф-ции: {min_value_3d} \nПри параметрах {min_params_3d}")
```
