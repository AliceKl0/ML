4. (2 балла) Обучите модель логистической регрессии на рассматриваемых данных из библиотеки sklearn. Рассчитайте метрики accuracy и выведите таблицу классификации.

```python
# Перед работой нормализируем данные.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import time
```

```python
# Обучение модели
start_time = time.time()
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X, y)
train_time_log_reg = time.time() - start_time
```

```python
# Оценка модели
y_pred = log_reg.predict(X)
accuracy_log_reg = accuracy_score(y, y_pred)
print(f"Accuracy (Logistic Regression): {accuracy_log_reg}")
print(f"Время обучения: {train_time_log_reg:.4f} секунд")
print("Classification Report:\n", classification_report(y, y_pred, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```

Метрика accuracy:
Общая точность модели составляет 0.97. Высокое значение точности указывает на то, что модель хорошо справляется с задачей классификации в текущем состоянии.

Метрика macro avg:
Средние значения для всех классов 0.97, что указывает на высокую общую производительность.

Метрика weighted avg:
Учитывая сбалансированность классов, взвешенная средняя точность равна 0.97, что подтверждает высокую производительность.

Confusion Matrix:

Classification Report:
Precision, Recall, и F1-Score не сильно варьируются между классами (метрики высоки, значения близки к 1).

5. (2 балла) Обучите полиномиальную модель классификации. Рассчитайте метрики accuracy и выведите таблицу классификации. Попробуйте разные степени полинома и выберите ту, которая работает лучше.

```python
from sklearn.preprocessing import PolynomialFeatures
import time
```

```python
degrees = [1, 2, 3, 4, 5]  # Пробуем разные степени полинома
best_degree = None
best_accuracy = 0
results_poly = []

for degree in degrees:
    # Преобразование данных с помощью PolynomialFeatures
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    # Создание и обучение модели логистической регрессии
    model = LogisticRegression(max_iter=1000)
    start_time = time.time()
    model.fit(X_poly, y)
    end_time = time.time()
    train_time = end_time - start_time
    results_poly.append(train_time)
    
    # Оценка точности
    y_pred = model.predict(X_poly)
    accuracy = accuracy_score(y, y_pred)
    
    print(f"Полином степени {degree}")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Время обучения: {train_time:.4f} секунд")
    print("Classification Report:")
    print(classification_report(y, y_pred, zero_division=0))
    print("Confusion Matrix:")
    print(confusion_matrix(y, y_pred))
    print("-" * 50)
    
    # Сохраняем лучшую степень
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_degree = degree

# Итог
print(f"Лучшая степень полинома: {best_degree} с accuracy {best_accuracy:.4f}")
```

Итак, лучшие результаты показывает модель со степенью полинома 2 (да, аналогичные результаты показывают и модели с полиномами 3, 4, 5).

6. (2 балла) Обучите модель классификации по методу опорных векторов. Рассчитайте метрики accuracy и выведите таблицу классификации. Попробуйте разные ядерные функции и выберите ту, которая работает лучше.

```python
from sklearn.svm import SVC
```

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
best_kernel = None
best_accuracy_svc = 0
results = []

for kernel in kernels:
    start_time = time.time()
    svm = SVC(kernel=kernel)
    svm.fit(X, y)
    train_time = time.time() - start_time

    y_pred = svm.predict(X)
    accuracy = accuracy_score(y, y_pred)
    results.append((kernel, accuracy, train_time))

    if accuracy > best_accuracy_svc:
        best_kernel = kernel
        best_accuracy_svc = accuracy

    print(f"Kernel: {kernel}, Accuracy: {accuracy}, Time: {train_time:.2f}s")

print(f"\nBest Kernel: {best_kernel}, Best Accuracy: {best_accuracy_svc}")
```

Итак, лучшие результаты показывает модель с ядром rbf (общая точность модели примерно 0.97)

7. (1 балл) Обучите модель классификации Перцептрон. Рассчитайте метрики accuracy и выведите таблицу классификации. 

```python
from sklearn.linear_model import Perceptron
```

```python
start_time = time.time()
perceptron = Perceptron(max_iter=1000)
perceptron.fit(X, y)
train_time_perc = time.time() - start_time

y_pred = perceptron.predict(X)
accuracy = accuracy_score(y, y_pred)

print(f"Accuracy (Perceptron): {accuracy}")
print(f"Время обучения: {train_time_perc:.4f} секунд")
print("Classification Report:\n", classification_report(y, y_pred, zero_division=0))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```

Общая точность модели составляет 0.87. Весьма высокое значение точности указывает на то, что модель хорошо справляется с задачей классификации в текущем состоянии.

Метрика macro avg: Средние значения не меньше 0.86, что указывает на высокую общую производительность.

Метрика weighted avg: Учитывая сбалансированность классов, взвешенная средняя точность не меньше 0.86, что подтверждает высокую производительность.

Confusion Matrix: Матрица показывает, что модель всё же далека от идеальной.

Classification Report: Precision, Recall, и F1-Score средне варьируются между классами.
