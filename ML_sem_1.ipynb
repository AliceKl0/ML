2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def func_2(n, f_el, diff_):
    arithm_prog = np.array([f_el + i * diff_ for i in range(n)])
    
    return np.diag(arithm_prog)
```

```python
# Пример работы ф-ции

n = 4
f_el = 1.5
diff_ = 3.5

func_2(n, f_el, diff_)
```

При помощи сетки графиков PairGrid визуализируйте попарные отношения признаков Rooms, Square, HouseFloor, HouseYear, Price следующим образом: на диагонали - гистограммы, под диагональю - ядерные оценки плотности, над диагональю - диаграммы рассеяния. По результатам визуализации сделайте выводы

```python
# Создаем PairGrid
g = sns.PairGrid(df, 
                 vars=["Rooms", "Square", "HouseFloor"], 
                 height=3)

# Диагональ: Гистограммы
g.map_diag(sns.histplot, kde=True, color='skyblue')

# Под диагональю: Ядерные оценки плотности
g.map_lower(sns.kdeplot, cmap="Blues", fill=True)

# Над диагональю: Диаграммы рассеяния
g.map_upper(sns.scatterplot, color='orange')

# Добавляем заголовок
plt.suptitle("Попарные отношения признаков", y=1.02)
plt.show()
```
