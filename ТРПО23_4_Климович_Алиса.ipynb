6. (4 балл) Обучите модель линейной регрессии на рассматриваемых данных, написанную своими руками. Проиллюстрируйте работу модели графиком обучения и распределения целевой переменной. Выведите уравнение полученной гиперплоскости.

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
scaler = MinMaxScaler()
cols = X.columns
X = scaler.fit_transform(X)
X = pd.DataFrame(data=X, columns=cols)
X
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class MultipleRegression(object):
    def __init__(self):
        self.b = None
        
    def predict(self, x):
        return x @ self.b
        
    def MSE(self, x, y):
        return (((y - self.predict(x)).T @ (y - self.predict(x))) / (2 * x.shape[0])).values
        
    def MAE(self, x, y):
        return (abs(y - self.predict(x)).mean()).values
        
    def MAPE(self, x, y):
        return (abs((y - self.predict(x))/y).mean()).values
        
    def coefs(self):
        return self.b
    
    def fit(self, x, y, alpha = 0.01, accuracy = 0.01, max_steps = 10000, intercept = True):
        y = np.array(y).reshape(-1, 1)
        
        if intercept:
            x['intercept'] = 1
        
        self.b = np.zeros((x.shape[1], 1))
        steps, errors = [], []
        step = 0
        
        for _ in range(max_steps):
            dJ_b = x.T @ (self.predict(x) - y) / x.shape[0]
            self.b -= alpha * dJ_b
            new_error = self.MSE(x, y)
            step += 1
            steps.append(step)
            errors.append(new_error)
        
        return steps, errors
```

```python
reg = MultipleRegression()
steps, errors = reg.fit(X, y)
predictions = reg.predict(X)

print("Предсказанные значения:", predictions)
```

```python
plt.scatter(predictions, y)
plt.plot(predictions, predictions, c='r')
```

```python
errors = np.array(errors).flatten()  # Преобразуем в одномерный масси

# Построение графика обучения
plt.plot(steps, errors)
plt.xlabel('Шаг')
plt.ylabel('Ошибка (MSE)')
plt.title('Ошибка обучения на каждом шаге')
plt.show()
```

```python
# Построение графика распределения целевой переменной
plt.scatter(range(len(y)), y, label='Реальные значения', color='blue')
plt.scatter(range(len(predictions)), predictions, label='Предсказанные значения', color='red')
plt.xlabel('Наблюдения')
plt.ylabel('Целевая переменная')
plt.legend()
plt.title('Реальные и предсказанные значения')
plt.show()
```

8. (1 балл) Оцените работу моделей, рассчитав для каждой из них метрики: коэффициент детерминации и ошибку MSE. Сделайте выводы по качеству работы моделей.

```python
# Коэффициент детерминации (R^2)
def r2_score(y_true, y_pred):
    ss_res = ((y_true - y_pred) ** 2).sum(axis=0)
    ss_tot = ((y_true - y_true.mean()) ** 2).sum(axis=0)
    return 1 - (ss_res / ss_tot).values[0]
```
